```python
data["num-of-cylinders"].tail()
```

```python
#Converts the object datatype of the float data type
#data["num-of-cylinders"].astype(float)
data['num-of-cylinders'] = pd.to_numeric(data['num-of-cylinders'],errors = 'coerce')
```

### Filter, Sort and Group By

```python
data["price"].max()
```

```python
data["price"].mean()
```

```python
# let us check how many vehicles are there whose price more than 30000
data[data["price"]>30000]
```

```python
data[data["price"]>30000].shape
```

```python
# let us check how many vehicles are there whose price more than 20000 and less than 30000

data[(data["price"]<30000) & (data["price"]>20000)]
```

```python
# Sorts values by col1 in ascending order
data.sort_values("price",ascending =False)
```

```python
# Sorts values by col1 in ascending order then col2 in descending order

data.sort_values(["average-mileage","price"],ascending = (True,False))
```

```python
# Returns a groupby object for values from one column

data.groupby(["company"]).sum()
```

```python
data.groupby(["company"])["price","average-mileage"].sum()
```

```python
data.groupby(["company","num-of-cylinders"],as_index=False)["price"].sum()
```

```python
data.groupby(["company"])["price","average-mileage"].sum()
```

```python
# Creates a pivot table that groups by col1 and calculates the sum of col2 and col3

data.pivot_table(index=["company"],values=["average-mileage","price"])
```

```python
# Finds the average across all columns for every unique column 1 group

data.groupby(["company"]).agg(np.mean)
```

```python
# Applies a function across each column

data.apply(np.min)
```

```python
# Applies a function across each row
data.apply(np.max,axis=0)
```

## Joining and Combining
Methods for combining two dataframes:


Creating dataframe 2 to perform joining operations

```python
# Adds the rows in df1 to the end of df2 (columns should be identical)
d2={'index' : [89,90] , 'company' : ['audi','bmw'] , 'body-style': ['wagon','sedan'], 'wheel-base':[112,98] , 'length':[154,176], 'engine-type':['dohc','rotor'] , 'num-of-cylinders':[5,6] , 'horsepower':[100,113], 'average-mileage':[67,54], 'price':[14567,12345]}
```

```python
d2
```

```python
data2 = pd.DataFrame(data=d2)
data2
```

```python
# Adds the rows in df1 to the end of df2 (columns should be identical)

data.append(data2)
```

```python
# Adds the columns in df1 to the end of df2 (rows should be identical)
pd.concat([data,data2], axis=1)
```
